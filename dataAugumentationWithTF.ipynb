# Importing Libraries

```python
import tensorflow as tf
import matplotlib.image as mpimg
import matplotlib.pyplot as plt
import numpy as np
from skimage.util import random_noise
from my_functions import show_image
```

# Importing original picture

```python
image_path = './output/train/altavoces/altavoces0.jpg'
image_string = tf.io.read_file(image_path, name=None)
image=tf.image.decode_jpeg(image_string,channels=3)


```

# Saturation

```python
satu_image = tf.image.adjust_saturation(image, 5)
show_image(image, satu_image, 'Adjust Saturation')
```

# Hue

```python
hue_image = tf.image.adjust_hue(image, delta=0.4)
show_image(image, hue_image, "Adjust HUE")
```

# Contrast

```python
constrast_image = tf.image.adjust_contrast(images=image, contrast_factor=0.6)
show_image(image, constrast_image, "Adjust Constrast")
```

# Brightness

```python
bright_image = tf.image.adjust_brightness(image, 0.2)
show_image(image, bright_image, "Adjust Brightness")
```

# Grayscale

```python
gray_image = tf.image.rgb_to_grayscale(image)
show_image(image, tf.squeeze(gray_image), "Grayscale image")
```

# Noise

```python
image_array = np.asarray(image)
noise_image = random_noise(image_array, mode="gaussian", var=0.1)
show_image(image, noise_image, "Noise Image")
```
